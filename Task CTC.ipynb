##### 1) Your views about the problem statement?
I believe the independent variables provided in the dataset are not very helpful. Even though there are quite a few of them, I think the model's accuracy could be significantly improved if the dataset included more variables that better explain the variation in the target variable (CTC).

##### 2) What will be your approach to solving this task?
My approach will follow the standard process. First, I will prepare the data to make it more suitable for modeling by performing steps like missing value imputation, variable transformation, and categorical variable encoding. Next, I will split the data into training and testing sets. Depending on the nature of the target variable, I will apply either regression models or classification models.

##### 3) What ML model options did you have available to perform this task?
Since the target variable (CTC) is numeric, I will start with linear regression models to establish a baseline. I will then move on to regularization techniques like Ridge and Lasso regression to improve performance and handle multicollinearity. After that, I will explore decision tree models, followed by pruning techniques, and finally, I will use ensemble methods to further enhance accuracy.

##### 4) Which models performance is best and what could be the possible reason for that?
I think Random Forest (with Grid Search) is the best model due to its high train and test scores. Moreover, the values are very close, which reduces the chances of overfitting and indicates good generalization.Comparing it to linear models, they have low r2 scores overall, and since all of them have lower test scores compared to train scores, this indicates underfitting or that the models couldn't adequately capture the complexity of the data. Now, comparing it to non-linear models, some of them, like boosting techniques (without grid search), indicates overfitting. On the other hand, some non-linear models have performed well but still fall short compared to Random Forest (with Grid Search).

##### 5) What steps can you take to improve this selected models performance even further?
It can be improved by using a wider range of hyperparameters, such as max_depth, min_samples_split, and max_features, to increase the accuracy of the model.
